```python
#Data Analysis Project on TWO_CENTURIES_OF_UM_RACES dataset
```

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("TWO_CENTURIES_OF_UM_RACES.csv")

#See the data is imported ?

df.head()
```

```python
#Check how many rows and columns are present in the dataset

df.shape
```

```python
#Check the metadata of the dataset

df.dtypes
```

```python
#In this project we are only considering USA races, 50Km/50Miles records within 2020 year
#Step1 - Filter out only 50Miles and 50Km records

df[(df["Event distance/length"].isin(['50mi','50km'])) & (df["Year of event"] == 2020)]
```

```python
#spliting only USA records 

df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA']
```

```python
#Combine all the filters

df2 = df[((df["Event distance/length"].isin(['50mi','50km'])) & (df["Year of event"] == 2020)) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2.head()
```

```python
#Cleaning the Data
#Remove USA from event name column

df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head()
```

```python
#Clean up athlete age
```

```python
df2['athlete_age'] = 2020 - df2['Athlete year of birth']
```

```python
df2.head()
```

```python
#Removing h from Athlete performance

df2['Athlete performance'] = df2['Athlete performance'].str.split('h').str.get(0)
```

```python
df2.head()
```

```python
#Dropping the columns 
df2 = df2.drop(['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis = 1)
```

```python
df2.head()
```

```python
#Cleaning up NULL values
df2.isna().sum()
df2 = df2.dropna()
```

```python
#Check Duplicate records

df2[df2.duplicated() == True]
```

```python
#Resetting index

df2.reset_index(drop = True)
```

```python
#Fixing Data types
df2.dtypes
```

```python
df2['athlete_age'] = df2['athlete_age'].astype(int)
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
df2.head()
```

```python
#Rename the Columns names

df2 = df2.rename(columns = {'Year of event': 'Year',
                            'Event dates': 'race_day',
                            'Event name': 'race_name',
                            'Event distance/length': 'race_length',
                            'Event number of finishers': 'number_of_finishers',
                            'Athlete performance': 'athelete_performance',
                            'Athlete gender': 'athlete_gender',
                            'Athlete ID': 'athelete_id',
                            'Athlete average speed': 'Athlete_average_speed'})
```

```python
df2.head()
```

```python
#sorting the columns

df3 = df2[['race_day', 'race_name', 'race_length', 'number_of_finishers', 'athelete_id', 'athlete_gender', 'athlete_age', 'athelete_performance', 'Athlete_average_speed']]
```

```python
#find two races from the records
df3.head()
```

```python
#Comparing number of people partispited in 50Mi and 50KM race in males and females

sns.histplot(df3, x = 'race_length', hue = 'athlete_gender')
```

```python
#display the distribution of average speeds of athletes in races with a length of 50 miles. 
#This visualization helps in understanding the variability and central tendency of athlete speeds in races of 
#this specific length*/
sns.displot(df3[df3['race_length'] == '50mi']['Athlete_average_speed'])
```

```python
sns.lmplot(data=df3, x= 'athlete_age', y= 'Athlete_average_speed')
```

```python
sns.lmplot(data=df3, x= 'athlete_age', y= 'Athlete_average_speed', hue= 'athlete_gender')
```

```python
# Speed for 50km, 50mi - male to female
df3.groupby(['race_length', 'athlete_gender'])['Athlete_average_speed'].mean()
```

```python
#Which age groups are the best in the 50mi race(20+race min)
df3.query('race_length == "50mi"').groupby('athlete_age')['Athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending = False).query('count>19')
```

```python
#Seson for the data
#spring 3-5
#Summer 6-8
#fall 9-11
#Winter 12-2

df3['race_month'] = df3['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df3['race_season'] = df3['race_month'].apply(lambda x: 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring')
```

```python
df3.groupby('race_season')['Athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending = False)
```
